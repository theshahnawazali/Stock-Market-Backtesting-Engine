```python
import numpy as np
import csv
import os
import math
```

```python
file_path = "HistoricalQuotes.csv"

data = []

if not os.path.exists(file_path):
    print("Dataset file Not found")
else:
    with open(file_path,"r") as f:
        reader = csv.reader(f)
        
        next(reader)
        
        for row in reader:
            data.append(row)

data = np.array(data)
```

```python
clean_data = []

for row in data:
    new_row = [
        row[0],  # Date same rahega
        float(row[1].replace('$','').strip()),
        int(row[2].strip()),
        float(row[3].replace('$','').strip()),
        float(row[4].replace('$','').strip()),
        float(row[5].replace('$','').strip())
    ]
    clean_data.append(new_row)

clean_data = np.array(clean_data)
print(clean_data)
```

```python
clean_data = []

for row in data:
    new_row = [
        float(row[1].replace('$','').strip()),
        int(row[2].strip()),
        float(row[3].replace('$','').strip()),
        float(row[4].replace('$','').strip()),
        float(row[5].replace('$','').strip())
    ]
    clean_data.append(new_row)

clean_data = np.array(clean_data)
clean_data = clean_data[::-1]
print(clean_data)
```

```python
close_price = clean_data[:, 0]
print(close_price)
open_prices = clean_data[:, 2]
high_prices = clean_data[:, 3]
low_prices = clean_data[:, 4]
volume = clean_data[:, 1]
```

```python
ma_short = np.convolve(close_price, np.ones(10)/10, mode='valid')
print(ma_short)
ma_long = np.convolve(close_price, np.ones(50)/50, mode='valid')
print(ma_long)
ma_short = ma_short[-len(ma_long):]
print(ma_long.size)
print(ma_short.size)
```

```python
signals = np.where(ma_short > ma_long, 1,
          np.where(ma_short < ma_long, -1, 0))

print(len(ma_short),len(ma_long),len(signals))
```

```python
cash = 10000
shares = 0
portfolio = []
trade = 0

for i in range(len(signals)):
    price = close_price[i]
    if signals[i] == 1: # Buy
        if cash != 0:
            print("Buying Shares..")
            shares = cash // price
            cash = 0
            print("Current Stock Price: ",price)
            print("Current Share: ", shares)
            print("Current Cash: ", cash)
            total = cash + shares * price
            portfolio.append(total)
            trade += 1
        
    elif signals[i] == -1:
        if shares != 0: # Sell
            print("Selling Shares..")
            cash = price * shares
            shares = 0
            print("Current Stock Price: ",price)
            print("Current Share: ", shares)
            print("Current Cash: ", cash)
            total = cash + shares * price
            portfolio.append(total)
            trade += 1
            
    else:
        print("Holding")
```

```python
print("Final Money:", portfolio[-1])
print("Profit:", portfolio[-1] - 10000)
return_per = ((portfolio[-1] - 10000)/ 10000) * 100
print(f"Return: {math.floor(return_per)}%")
print("Total Trade: ",trade)
```
